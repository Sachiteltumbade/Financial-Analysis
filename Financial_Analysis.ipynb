```python
import pandas as pd

# Read the data into a pandas DataFrame
data = pd.read_csv('/content/Financial Analytics data.csv')
data.head()
```

```python

from matplotlib import pyplot as plt
data.plot(kind='scatter', x='Mar Cap - Crore', y='Sales Qtr - Crore', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
#dropping missing values
data = data.dropna(subset=['Mar Cap - Crore', 'Sales Qtr - Crore'])
data = data[(data['Mar Cap - Crore'] >= 0) & (data['Sales Qtr - Crore'] >= 0)]
```

```python
#sum of Mar cap crore and sales qtr crore
sum_products = sum((data['Mar Cap - Crore'] * data['Sales Qtr - Crore']).values)
```

```python
#sum of squares of mar cap crore
sum_mar_cap_squared = sum(data['Mar Cap - Crore']**2)
sum_sales_squared = sum(data['Sales Qtr - Crore']**5)
```

```python
#Mean of Mar Cap - Crore and Sales Qtr - Crore
mar_cap_mean = data['Mar Cap - Crore'].mean()
sales_mean = data['Sales Qtr - Crore'].mean()
```

```python
#Covariance
covariance = sum_products / len(data) - mar_cap_mean * sales_mean
```

```python
#standard deviations
mar_cap_std = (sum_mar_cap_squared / len(data) - mar_cap_mean**2)**0.5
sales_std = (sum_sales_squared / len(data) - sales_mean**2)**0.5
```

```python
correlation = covariance / (mar_cap_std * sales_std)
```

```python
print(f"Market Capitalization Mean:{mar_cap_mean} crores")
print(f"Sales Mean: {sales_mean} crores")
print(f"Correlation between Market Capitalization and Sales: {correlation}")
```

```python
top_10_market_cap = data.nlargest(10, 'Mar Cap - Crore')
top_10_sales = data.nlargest(10, 'Sales Qtr - Crore')

print("Top 10 Companies by Market Capitalization:")
print(top_10_market_cap[['Name', 'Mar Cap - Crore']])

print("Top 10 Companies by Sales:")
print(top_10_sales[['Name', 'Sales Qtr - Crore']])
```
